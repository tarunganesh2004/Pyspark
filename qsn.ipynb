```python
from pyspark import SparkContext

sc=SparkContext(master="local",appName="test1")

df=sc.textFile("students.txt")

df.collect()
```

```python
df_split=df.map(lambda x:x.split(","))
df_split.collect()
```

```python
df_formatted=df_split.map(lambda x:(x[0][:4],x[0][4:],x[1],*x[2].split(":")))
df_formatted.collect()
```

```python
# total Marks
total_marks=df_formatted.map(lambda x:(int(x[3])+int(x[4])+int(x[5])+int(x[6]))/4)
total_marks.collect()



Grade=total_marks.map(grade)
Grade.collect()
```

```python
def grade(avg):
    if avg>=80:
        return "A"
    elif avg>=60:
        return "B"
    elif avg>=40:
        return "C"
    else:
        return "F"
df_final=df_formatted.map(lambda x:x+((int(x[3]) + int(x[4]) + int(x[5]) + int(x[6])) ,)).map(lambda x:x+(grade(x[7]/4),))

df_final.collect()
```

```python
df_final.filter(lambda x:x[8]=="A" or x[8]=="B").collect()
```

```python
header = ("number","name","gender","maths","physics","biology","chemistry","total_marks","grade")

header_rdd=sc.parallelize([header])

res=header_rdd.union(df_final)
# res.collect()
for row in res.collect():
    print(",".join(map(str,row)))
```
